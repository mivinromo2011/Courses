```python
import numpy as np
from matplotlib import pyplot as plt
import pandas as pd
import seaborn as sb
from scipy.stats import iqr
```

```python
data=pd.read_csv(r'heart.csv')
```

```python
print(data.shape)

print("==========================================================")

print("Info")
print("------")
print(data.info)
```

```python
print(data.isna().sum())
```

```python
print(data.mean())
```

```python
print("Mode")
print("------")
print(data.mode())
print("==========================================================")
print("Median")
print("------")
print(data.median())
print("==========================================================")
print("Standar Deviation")
print("------------------")
print(data.std())
print("==========================================================")
print("Variance")
print("----------")
print(data.var())
print("==========================================================")
```

```python
print("Inter Quartile Range")
print("--------------------")
for(cname,cvalue) in data.iteritems():
	print(cname,' : ',iqr(cvalue.values))
```
